```python
data=pd.read_csv('train_data.csv')
print(data.shape)
X=data.iloc[0:2000,0:4]
Y=data.iloc[0:2000,4:5]
datatest=pd.read_csv('test_data.csv')
X_test=datatest.iloc[0:201,0:4]
print(datatest.shape)
```

```python
from sklearn import preprocessing as pre
scaler = pre.StandardScaler().fit(X)
X_t_s = scaler.transform(X)
X_test_s = scaler.transform(X_test)
print(X_test_s)
```

```python
SVR_model = svm.SVR(kernel='rbf',C=100000,gamma=.01).fit(X,Y)
```
